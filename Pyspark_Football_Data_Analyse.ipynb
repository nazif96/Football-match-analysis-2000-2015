```python
# importation des données
df_matches= spark.read.format('csv').options(header = 'True').load('/content/drive/MyDrive/Colab Notebooks/Data/football_matches.csv')

# preview des données

df_matches.limit(20).show()
```


**Signification des caractéristiques**

- Match_ID : Identifiant unique pour chaque match
- Div : Niveau du championnat (D1 = premier niveau, D2 = deuxième niveau, E0 = troisième niveau)
- Season : Année de début de la saison au cours de laquelle le match s'est déroulé
- Date : Date du match
- HomeTeam : équipe jouant dans son stade
- AwayTeam : équipe jouant dans le stade de l'équipe adverse
- FTHG : Nombre de buts de l'équipe locale
- FTAG : Nombre de buts de l'équipe adverse
- FTR : Résultat final

Traduit avec DeepL.com (version gratuite)

```python
# Renommer les colonnes

df_matches = df_matches.selectExpr(
    "*",
    "`FTHG` AS `HomeTeamGoals`",
    "`FTAG` AS `AwayTeamGoals`",
    "`FTR` AS `FinalResult`"
)
```

```python
# preview des données

df_matches.limit(20).show()
```

```python
# supprimer les colonnes

df_matches = df_matches.drop("FTHG", "FTAG", "FTR")

# preview des données

df_matches.limit(10).show()
```

```python
df_matches.toPandas().head(10)
```

```python
# Création des colonnes binaires pour la victoire , defaites et nuls

#df_matches = df_matches.withColumn("HomeTeamWin", df_matches["FinalResult"] == "H")
df_matches = df_matches.withColumn("HomeTeamWin", when(col("FinalResult") == "H", 1).otherwise(0))

# preview des données
df_matches.limit(10).show()
```

```python
df_matches = df_matches.withColumn("AwayTeamWin", when(col("FinalResult") == "A", 1).otherwise(0)) \
                        .withColumn("Gametie", when(col("FinalResult") == "D", 1).otherwise(0))

# preview des données
df_matches.limit(10).show()
```

Filtrage

Création de dataframe contenant que des matches de bundesliga

```python
df_bundesliga = df_matches.filter((col("Div") == "D1") &
                        (col("Season") >= 2000) &
                        (col("Season") <= 2015))

# preview des données
df_bundesliga.limit(10).show()
```

```python
df_bundesliga.toPandas()
```

```python
# Création de dataframe contenant les statistiques des matches à domicile

df_home_matches= df_bundesliga.groupBy('Season', 'HomeTeam') \
    .agg(sum('HomeTeamWin').alias('TotalHomeWin'),
         sum('AwayTeamWin').alias('TotalHomeLoss'),
         sum('Gametie').alias('TotalHometie'),
         sum('HomeTeamGoals').alias('HomeScoreGoals'),
         sum('AwayTeamGoals').alias('HomeAgainstGoals')) \
         .withColumnRenamed('HomeTeam', 'Team')


  # preview des données
df_home_matches.toPandas()
```

```python
# Création de dataframe contenant les statistiques des matches à extérieur

df_away_matches= df_bundesliga.groupBy('Season', 'AwayTeam') \
    .agg(sum('AwayTeamWin').alias('TotalAwayWin'),
         sum('HomeTeamWin').alias('TotalAwayLoss'),
         sum('Gametie').alias('TotalAwaytie'),
         sum('AwayTeamGoals').alias('AwayScoreGoals'),
         sum('HomeTeamGoals').alias('AwayAgainstGoals')) \
         .withColumnRenamed('AwayTeam', 'Team')


  # preview des données
df_away_matches.toPandas()
```

JOINTURE

```python
# join home & away data
df_merge = df_home_matches.join(df_away_matches, ['Team','Season'], 'inner')

# preview des données
df_merge.toPandas()
```

```python
# Création de colonnes pour les buts et resultats

df_totals = df_merge.withColumn('GoalsScored', col('HomeScoreGoals') + col('AwayScoreGoals')) \
                    .withColumn('GoalsAgainst', col('HomeAgainstGoals') + col('AwayAgainstGoals')) \
                    .withColumn('Win', col('TotalHomewin') + col('TotalAwayWin')) \
                    .withColumn('Loss', col('TotalHomeLoss') + col('TotalAwayLoss')) \
                    .withColumn('Tie', col('TotalHometie') + col('TotalAwaytie'))

# preview des données
df_totals.toPandas()
```

```python
# suppression des colonnes non necessaire

cols_to_drop = ['HomeScoreGoals', 'AwayScoreGoals', 'HomeAgainstGoals', 'AwayAgainstGoals', 'TotalHomeWin',
                 'TotalAwayWin', 'TotalHomeLoss', 'TotalAwayLoss','TotalHometie', 'TotalAwaytie']

# suppression
df_cleaned = df_totals.drop(*cols_to_drop)

# preview des données
df_cleaned.toPandas()
```

```python
#colonnes additionelles

df_processed = df_cleaned.withColumn('GoalDifferentials', col('GoalsScored')- col('GoalsAgainst')) \
                         .withColumn('WinPercentage', round((100* col('Win') / (col('Win')+col('Loss')+ col('Tie'))),2))

# preview des données
df_processed.toPandas()
```

```python
from pyspark.sql.window import Window
#from pyspark.sql.functions import row_number
```

```python
# set window for partition

window_partition = Window.partitionBy('Season').orderBy(col('WinPercentage').desc(), col('GoalDifferentials').desc())

# rank team by season
df_ranked = df_processed.withColumn('TeamPosition', rank().over(window_partition))

# preview des données
df_ranked.toPandas()
```

```python
# tableau pour afficher le premier de chaque saison

df_top_teams = df_ranked.filter(col('TeamPosition') == 1)

# preview des données
df_top_teams.toPandas()
```
